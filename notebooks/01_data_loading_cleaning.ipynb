```python
import pandas as pd

df = pd.read_csv("../data/raw/superstore_clean.csv")
df.head()

df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")

df[["Order Date", "Ship Date"]].head()

df.info()
```

```python
df["Sales"] = (
    df["Sales"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(".", "", regex=False)
    .astype(float)
)

df["Profit"] = (
    df["Profit"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(".", "", regex=False)
    .astype(float)
)
```
